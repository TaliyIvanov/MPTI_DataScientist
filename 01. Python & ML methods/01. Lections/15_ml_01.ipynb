## Решим модельную задачку классификации

```python
# сгенерируем данные с помощью .make_classification()
# n_samples - кол-во наших сэмплов
# n_features - кол-во фич
# n_informative -
# n_classes - кол-во классов, которые мы нагенерировали
# n_redundant -
# n_clusters_per_class - кол-во кластеров на каждый класс
# class_sep - 
problem = datasets.make_classification(n_samples=200,
                                       n_features=2, 
                                       n_informative=2, 
                                       n_classes=2, 
                                       n_redundant=0, 
                                       n_clusters_per_class=1,
                                       class_sep=0.1)

colors = ListedColormap(['green', 'red'])

draw(problem, colors)
```

### Разобьем данные на обучающую и тестовые выборки

```python
problem
```

```python
# импортируем модуль для разбиения наших данных
from sklearn import model_selection
# train_data - тренировочные данные
# test_data - тестовые данные
# train_labels - тренировочные лейблы
# test_labels - тестовые лейблы
# problem[0] - передаем наши x и y из первого массива
# problem[1] - передаем ответы из второго массива
# test_size=0.3 - 

train_data, test_data, train_labels, test_labels = model_selection.train_test_split(problem[0], 
                                                                                    problem[1], 
                                                                                    test_size=0.3)
```

```python
# отрисуем наши тренировочные данные 
draw((train_data, train_labels), colors)
```

```python
# отрисуем тестовые данные
draw((test_data, test_labels), colors)
```

### Импортируем модель kNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# присвоим модель переменной
# в данной модели была использована метрика Minkowski (среднее квадратичное отклонение)
model = KNeighborsClassifier()
```

```python
model
```

### Проверим качество модели сначала на обучающей выборке

```python
# импортируем метрики для оценки качество
from sklearn.metrics import accuracy_score
```

```python
# тренировочные предсказания модели
train_predicts = model.predict(train_data)
# предсказания наей модели для каждого из примеров
train_predicts
```

```python
draw((train_data, train_predicts), colors)
```

```python
draw((train_data, train_labels), colors)
```

```python
# оценим качество нашей простой модели с количеством соседеней =5
accuracy_score(train_labels, train_predicts)
```

```python
# если количество соседей равно k = 1
model_k1 = KNeighborsClassifier(n_neighbors=1)
model_k1.fit(train_data, train_labels)
train_predicts_k1 = model_k1.predict(train_data)
assert np.all(np.isclose(train_labels, train_predicts_k1))
accuracy_score(train_labels, train_predicts_k1)
```

### Теперь сделаем то же самое на тестовых данных

```python
# k = 5
test_predicts = model.predict(test_data)
accuracy_score(test_labels, test_predicts)
```

```python
# k = 1
test_predicts_k1 = model_k1.predict(test_data)
accuracy_score(test_labels, test_predicts_k1)
```

### Дискуссия


1.   Хороший или плохой это результат?
2.   Что можно сделать по другому?



```python
# сравним для разного количества соседей от 1 до 20
# разумеется это все для процесса обучения, в реально жизни обучить 20 моделей
# долго и, соответственно, дорого
d = {}  # k --> acc
for k in range(1, 21):
    model_ = KNeighborsClassifier(n_neighbors=k)
    model_.fit(train_data, train_labels)
    test_predicts_ = model_.predict(test_data)
    d[k] = accuracy_score(test_labels, test_predicts_)
    
print(*d.items(), sep='\n')
```

Лучшая accuracy у 3 и 5 соседей

#### Визуализация
Вообще первый этап построения модели это всегда исследовательский анализ данных. Он дает нам базовое представление о данных, их разнообразии, наличии ошибок и выбросов, корреляций. Кроме того сам формат и распределение данных при понимании внутренней кухни различных моделей формирует вектор исследования, с каких моделей начать, что сработает, а что скорее всего нет, какие могут быть проблемы на этапе разбиения данных и т.д. Визуализация на этапах улучшения качества формирует представление о том, на каких примерах текущая итерация модели ошибается. Благодаря этому можно понять чего не хватает или как нужно трансформировать данные, чтобы модель учла свои ошибки на следующей итерации цикла создания модели.

## Кросс валидация

Кросс валидация - разбиаения датасет на несколько частей.
Далее обучаем и тестируем меняя эти части

```python
# переименуем переменные. чтобы можно было из документации копировать - вставлять и оно работало

X_train = train_data
y_train = train_labels
X_test = test_data
y_test = test_labels
```

```python
# загрузим необходимые модули
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
```

```python
# В sklearn есть специальный модуль для работы с кросс-валидацией
from sklearn.model_selection import GridSearchCV

# Будем использовать модель k ближайших соседей, которую проходили на 5 занятии
from sklearn.neighbors import KNeighborsClassifier

# Создадим классификатор
knn = KNeighborsClassifier()

# Зададим сетку - среди каких значений выбирать наилучший параметр.
knn_grid = {'n_neighbors': np.array(np.linspace(1, 25, 25), dtype='int')} # перебираем по параметру <<n_neighbors>>, по сетке заданной np.linspace(2, 100, 10)

# Создаем объект кросс-валидации
gs = GridSearchCV(knn, knn_grid, cv=5, return_train_score=True)

# Обучаем его
gs.fit(X_train, y_train)
```

```python
# Функция отрисовки графиков

def grid_plot(single_param_range, train_results, test_results, x_label, title, y_label):
    plt.figure(figsize=(12, 6))
    plt.grid(True)
    plt.plot(single_param_range, train_results, 'go-', label="Train")
    plt.plot(single_param_range, test_results, 'bo-', label="Test")
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.legend()
    plt.title(title)
```

```python
gs.cv_results_.keys()
```

```python
# Строим график зависимости качества от числа соседей
# замечание: результаты обучения хранятся в атрибуте cv_results_ объекта gs

grid_plot(knn_grid['n_neighbors'], gs.cv_results_['mean_train_score'], gs.cv_results_['mean_test_score'], 'n_neighbors', 'KNeighborsClassifier', 'accuracy')
```

А теперь повторим эту процедуру много раз, и произведем подбор наилучшего множества гиперпараметров модели.

```python
# функция, производящая подбор гиперпараметров модели на кросс валидации, а также оценку на тестовой выборке
# принимает на вход объект классификатора
# словарь с диапазонами изменений гиперпараметров модели
# функцию для подсчета метрики или ее название среди доступных метрик для GridSearchCV (или список функций и/или названий) https://scikit-learn.org/stable/modules/model_evaluation.html#scoring-parameter
# и n разбиений данных на обучение-валидацию (n фолдов) или объект "разбивателя" из модуля sklearn.model_selection или же просто число разбиений
# также у вас должны быть определены переменные X_train, y_train и X_test, y_test
def perform_cv(base_classifier, tuned_parameters, score, cv):
  print("# Tuning hyper-parameters for %s" % score)
  print()

  clf = GridSearchCV(base_classifier, tuned_parameters, cv=cv,
                      scoring=score)
  clf.fit(X_train, y_train)

  print("Best parameters set found on development set:")
  print()
  print(clf.best_params_)
  print()
  print("Grid scores on development set:")
  print()
  means = clf.cv_results_['mean_test_score']
  stds = clf.cv_results_['std_test_score']
  for mean, std, params in zip(means, stds, clf.cv_results_['params']):
      print("%0.3f (+/-%0.03f) for %r"
            % (mean, std * 2, params))
  print()

  print("Detailed classification report:")
  print()
  print("The model is trained on the full development set.")
  print("The scores are computed on the full evaluation set.")
  print()
  y_true, y_pred = y_test, clf.predict(X_test)
  print(classification_report(y_true, y_pred))
  print()
  return clf
```

```python
knn = KNeighborsClassifier()
```

```python
tree = DecisionTreeClassifier()
```

```python
bayes = GaussianNB()
```

```python
from sklearn.model_selection import KFold
skf = KFold(n_splits=5) # разбиение на 5 фолдов
```

```python
gs = perform_cv(knn, {'n_neighbors': [2,3,4,5,6,7]}, 'accuracy', skf)
```

```python
gs = perform_cv(tree, {'max_depth': [2,3,4,5]}, 'accuracy', skf)
```
